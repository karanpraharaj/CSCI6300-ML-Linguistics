##### Write a snippet of code to figure out the following:
- What's the most frequent stress pattern for disyllabic nouns? 
- disyllabic verbs? 
- disyllabic adjectives?

```python
stress_pattern_freq = {}

for line in lines:
    entry = line.split('\t')
    num_syllables = entry[2]
    pos_all = entry[5].strip('()').split()
    stress_pattern = entry[1]
    
    if num_syllables not in stress_pattern_freq:
        stress_pattern_freq[num_syllables] = {'V':{}, 'A':{}, 'N':{}}
    
    for pos in pos_all:
        if pos[0] in {'V', 'A', 'N'} and pos!='AV':
            if stress_pattern in stress_pattern_freq[num_syllables][pos[0]]:
                stress_pattern_freq[num_syllables][pos[0]][stress_pattern] += 1
            else:
                stress_pattern_freq[num_syllables][pos[0]][stress_pattern] = 1
        else:
            continue
```

```python
stress_pattern_freq
```

```python
print("The most frequent stress pattern for disyllabic (S2) nouns (N) = ", max(stress_pattern_freq['S2']['N'] , key=stress_pattern_freq['S2']['N'].get))
print("The most frequent stress pattern for disyllabic (S2) verbs (VB, VT) = ", max(stress_pattern_freq['S2']['V'] , key=stress_pattern_freq['S2']['V'].get))
print("The most frequent stress pattern for disyllabic (S2) adjectives (AJ) = ", max(stress_pattern_freq['S2']['A'] , key=stress_pattern_freq['S2']['A'].get))
```

##### Write a function that takes a string of words as arguments (space-separated), and returns the pronunciation:

```python
def pronounce(sentence):
    pronunciation = ""
    words = sentence.split()
    for word in words:
        pronunciation = pronunciation + pron(word)+ " "
    
    return pronunciation
```
