Write a function that (1) takes a set/list of letters as input and a list, and (2) returns a new list where all words that contain any of the letters on the list is removed. It's a good idea to test your code with a short list such as the one above before testing it on the actual wordle-list.

```python
def removeLetters(myList, letterList):
    revisedList = []
    for word in myList:
        flag = 0
        for i in letterList:
            if i in word:
                flag = 1
                break
        if flag == 1:
            continue
            
        revisedList.append(word)
        
    return revisedList
```

```python
mylist = ['catch', 'doggo', 'horse', 'mouse']  # Small list to test the function.
myletters = ['a', 'u']
print(removeLetters(mylist, myletters))
```

```python
# Testing on Wordle list

myletters = ['a', 'u','i','e','o']
print(removeLetters(wordlelist, myletters))
```

Additionally (if you make it this far), write another function that takes a list of pairs (letter, position) and removes those words from the list that don't have the letter in the correct position.

```python
def removePosition(myList, pairsList):
    revisedList = []
    for word in myList:
        flag = 0 
        for pair in pairsList:
            if word[pair[1]] != pair[0]:
                flag = 1
                break
            
        if flag == 1:
            continue
        
        revisedList.append(word)
    
    return revisedList
        
```

```python
# Testing on Wordle list

mypairs = [('o', 1), ('e', 4)]
removePosition(wordlelist, mypairs)
```
